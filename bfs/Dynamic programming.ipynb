```python
from collections import deque

def compute_value(grid,goal,cost):
    # ----------------------------------------
    # insert code below
    # ----------------------------------------
    
    N = len(grid)
    M = len(grid[0])
    value = [[9 for col in range(len(grid[0]))] for row in range(len(grid))]
    visit = [[0 for col in range(len(grid[0]))] for row in range(len(grid))]
    
    que = deque([[goal[0], goal[1], 0]])
    
    while len(que) > 0:
        r, c, v = que.popleft()
        visit[r][c] = 1
        value[r][c] = v
        
        for dr, dc in delta:
            nr = r + dr
            nc = c + dc
            if (0 <= nr < N) and (0 <= nc < M) and visit[nr][nc] == 0 and grid[nr][nc] == 0:
                que.append([nr, nc, v+cost])

    return value
```

```python
def optimum_policy(grid,goal,cost):
    N = len(grid)
    M = len(grid[0])
    value = [[99 for col in range(len(grid[0]))] for row in range(len(grid))]
    visit = [[0 for col in range(len(grid[0]))] for row in range(len(grid))]
    policy = [['≈' for col in range(len(grid[0]))] for row in range(len(grid))]
    
    que = deque([[goal[0], goal[1], 0]])
    
    while len(que) > 0:
        r, c, v = que.popleft()
        visit[r][c] = 1
        value[r][c] = v
        for (dr, dc), d_name in zip(delta, delta_name):
            nr = r + dr
            nc = c + dc
            if (0 <= nr < N) and (0 <= nc < M) and visit[nr][nc] == 0 and grid[nr][nc] == 0:
                que.append([nr, nc, v+1])
       
    visit = [[0 for col in range(len(grid[0]))] for row in range(len(grid))]
    for r in range(N):
        for c in range(M):
            if grid[r][c] == 1:
                continue
            ways = []
            for (dr, dc), d_name in zip(delta, delta_name):
                nr = r + dr
                nc = c + dc
                
                if (0 <= nr < N) and (0 <= nc < M) and visit[nr][nc] == 0 and grid[nr][nc] == 0:
                    ways.append([value[nr][nc], d_name])
            
            if(len(ways) > 0):
                ways.sort()
                policy[r][c] = ways[0][1]
    policy[goal[0]][goal[1]] = '*'
    pprint(value)
    return policy
```
